## Coeficiente de Inteligência
Vamos comparar os QI entre diferentes grupos usando Seaborn

```python
#Importar as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # biblioteca espefica para fazer plotes(graficos), tem um design mais elegante.
import squarify
```

```python
#ler o DataFrame
mydata = pd.read_csv("scores.csv")
mydata.head()
```

```python
#vamos salvar as colunas do dataframe em diferentes listas

group_a = mydata[mydata.columns[0]].tolist()
group_b = mydata[mydata.columns[1]].tolist()
group_c = mydata[mydata.columns[2]].tolist()
group_d = mydata[mydata.columns[3]].tolist()
```

```python
#criar um novo dataframe com apenas duas colunas
data = pd.DataFrame({'Groups': ['Group A'] * len(group_a) \
                             + ['Group B'] * len(group_b) \
                             + ['Group C'] * len(group_c) \
                             + ['Group D'] * len(group_d),
                     
                     'IQ score': group_a + group_b + group_c + group_d})
```

```python
data.head()
```

```python
plt.figure(dpi=150)
# Set style
sns.set_style('whitegrid')
# Create boxplot
sns.boxplot('Groups', 'IQ score', data=data, showmeans = True)
# Despine
sns.despine(left=True, right=True, top=True, bottom = True)
#Add title
plt.title('QI para diferentes grupos')
# Show plot
plt.show()
```

## Gráfico Violino:

```python
# Create figure
plt.figure(dpi=150)
# Set style
sns.set_style('whitegrid')
# Create boxplot
sns.violinplot('Groups', 'IQ score', data=data)
# Despine
sns.despine(left=True, right=True, top=True)
# Add title
plt.title('QI para diferentes grupos')
# Show plot
plt.show()
```

## Gráfico de Barras

```python
mydata = pd.read_csv("movie_scores.csv")
mydata.head()
```

```python
# Criando o DataFrame
movie_scores = pd.DataFrame({"Movie Title": list(mydata["MovieTitle"]) * 2,
                "Score": list(mydata["AudienceScore"]) + list(mydata["Tomatometer"]),
                "Type": ["Audience Score"] * len(mydata["AudienceScore"]) + \
                             ["Tomatometer"] * len(mydata["Tomatometer"])})
movie_scores.head()
```

```python
sns.set()
plt.figure(figsize=(10, 5), dpi=300)
# Create bar plot
ax = sns.barplot("Movie Title", "Score", hue="Type", data=movie_scores)
plt.xticks(rotation=90)

# Add title
plt.title("Nota de Filmes em duas bases")
plt.xlabel("Filmes")
plt.ylabel("Pontuação de Filmes")
plt.ylim(0, 115)

# Show plot
plt.show()
```

## Outra forma de fazer o grafico acima com quebra de linha do eixo x

```python
# Parte adaptada do Edu para fazer a quebre do eixo x

def split_line(text):
    x = ''
    stop = 0
    for i in range(len(text)):
        j = i+1
        if ((i % 10) < 3) and (text[i] == " "):
            x = x + text[stop:j] + "\n" 
            stop = i
    if stop != 0:
        x = x + text[stop:]
    else:
        x = text
    return x

    #x_labels = movie_scores['Movie Title'].apply(lambda x: [x[i:i+n] for i in range(0, len(x), n)])
x_labels = movie_scores['Movie Title'].apply(lambda x: split_line(x))

# Codigo anterior do lucio
sns.set()
plt.figure(figsize=(15, 5))

ax = sns.barplot(x="Movie Title", y="Score", hue="Type", data=movie_scores)
#plt.xticks()

# Add title
plt.xticks(ticks = range(len(x_labels[0:5])), labels = x_labels[0:5])
plt.title("Nota de Filmes em duas bases")
plt.xlabel("Filmes")
plt.ylabel("Pontuação de Filmes")
plt.ylim(0, 115)

# Show plot
plt.show()
```

## Uma visão geral de distribuição em um quadrado

```python
mydata = pd.read_csv("water_usage.csv")
mydata.head()
```

```python
# Create figure
plt.figure(dpi=200)
# Create tree map
labels = mydata['Usage'] + ' (' + mydata['Percentage'].astype('str') + '%)'
squarify.plot(sizes=mydata['Percentage'], label=labels, color=sns.light_palette('green', mydata.shape[0]))

plt.axis('off')  #esconde as linhas dos eixos, off e para nao plotar nenhum eixo.
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
