### Colocar um gráfico ao lado do outro

```python
rank = sales.Rank[0:10]
nasales = sales.NA_Sales[0:10]
eusales = sales.EU_Sales[0:10]
jpsales = sales.JP_Sales[0:10]


plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.plot(rank, nasales)
plt.ylabel('Vendas', size = 20)
plt.legend(['NA'])

plt.subplot(132)
plt.plot(rank, eusales)
plt.xlabel('Ranking - Global', size = 20)
plt.legend(['Europa'])

plt.subplot(133)
plt.plot(rank, jpsales)
plt.legend(['Japão'])

plt.suptitle('Vendas por Regiões', size = 24)

plt.show()
```

### Hora de mudar um pouco a aparência dessa linha azul sem graça (não que vamos fazer algo de outro mundo)

```python
# Iniciar o plot
plt.plot(rank, gsales,  drawstyle='steps', color='red', linewidth=5)
# Alguns pontos de atenção:
#       1. Estamos informando como as linhas irão conectar os pontos
#       2. Estamos definindo a cor da linha
#       3. Estamos fixando a largura da linha

#nome dos eixo x e y
plt.xlabel('Posição no Rank')
plt.ylabel('Vendas Globais')

#informando o titulo do gráfico
plt.title('Vendas Globais de Jogos')

# definindo a legenda
plt.legend(['10 mais vendidos'])

#mandando mostrar o gráfico
plt.show;
```

### Para começar vamos fazendo algumas operações com o DataFrame
1. O primeiro gráfico mostrará o volume de venda de jogos por ano, para isso vamos:
    1. Criar o DataFrame (essa parte não é necessária, mas vamos fazê-la por questões pedagógicas)

## Gráfico de barras empilhado

### DataFrame - Vendas Regionais agrupadas por genêro

```python
genre = sales[[ "Genre", 'NA_Sales','EU_Sales','JP_Sales']].groupby(['Genre'], as_index=False).sum()
genre
```

### Drawing the stacked plot

```python
f, ax = plt.subplots(1, figsize=(12,5))

bar_l = genre['Genre']

ax.bar(bar_l, genre['NA_Sales'], label='NA', color='red')
ax.bar(bar_l, genre['EU_Sales'], bottom=genre['NA_Sales'], label='EU',color='blue')
ax.bar(bar_l, genre['JP_Sales'], bottom=[i+j for i,j in zip(genre['NA_Sales'],genre['EU_Sales'])], \
       label='Japan',color='green')

ax.set_title("Vendas por Gênero e Região")
ax.set_xlabel("Gênero");
ax.set_ylabel("Vendas");

plt.legend(loc='upper right');
```
