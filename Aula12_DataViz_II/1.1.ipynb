### Colocar um gráfico ao lado do outro

```python
rank = sales.Rank[0:10]
nasales = sales.NA_Sales[0:10]
eusales = sales.EU_Sales[0:10]
jpsales = sales.JP_Sales[0:10]


plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.plot(rank, nasales)
plt.ylabel('Vendas', size = 20)
plt.legend(['NA'])

plt.subplot(132)
plt.plot(rank, eusales)
plt.xlabel('Ranking - Global', size = 20)
plt.legend(['Europa'])

plt.subplot(133)
plt.plot(rank, jpsales)
plt.legend(['Japão'])

plt.suptitle('Vendas por Regiões', size = 24)

plt.show()
```

```python
# Outra forma de fazer o mesmo grafico

# crio uma variavel e mostro quantas linhas e quantos graficos tera e seu tamanho.
fig, ax= plt.subplots(1,3,figsize=(15, 5))

# chamo o primeiro grafico de indice zero.
ax[0].plot(rank, nasales)
ax[0].set_ylabel('Vendas', size = 20)
ax[0].legend(['NA'])

#chamando o segundo grafico de indice 1
ax[1].plot(rank, eusales)
ax[1].set_xlabel('Ranking - Global', size = 20)
ax[1].legend(['Europa'])

#chamando o terceiro grafico de indice 2
ax[2].plot(rank, jpsales)
ax[2].legend(['Japao'])

plt.suptitle('Vendas por Regiões', size = 24)

plt.show()
```

## Vamos plotar algumas barras agora

### Para começar vamos fazendo algumas operações com o DataFrame
1. O primeiro gráfico mostrará o volume de venda de jogos por ano, para isso vamos:
    1. Criar o DataFrame (essa parte não é necessária, mas vamos fazê-la por questões pedagógicas)
    2. Faremos um groupby por ano
    3. Fazer a soma da venda global

```python
#Trabalhando com o dataframe de vendas de jogos
# Quando  o indice nao esta no seu dataframe, voce precisa dar um  reset_index ou se o indice tiver 
# Caso o index ja esteja no seu dataframe, set_index ja basta.
# se usar o comando pd.rolling(window) ele te mostra quantas linhas vc esta andando por vez. Window vc define \
# o ano que sera somando, se e de 2 em 2 anos, 3 em 3  e etc.
vendas_anuais = pd.DataFrame({'total_global_sales' : sales.groupby( [ "Year"] )['Global_Sales'].
                             sum()}).reset_index().set_index(['Year'])
vendas_anuais.head()
```

### Gráfico de barras

```python
#definir o tamanho do gráfico
plt.figure(figsize=(12,5));

#mostrar o gráfico
# estou definindo somente a coluna que quero plotar e automaticamente meu indice sera o eixo X.
chart = vendas_anuais.total_global_sales.plot(kind='bar');
```

### Seria possível imprimir o mesmo gráfico sem necessariamente fazer um novo DataFrame

### Organize o volume de vendas de maneira descrescente, e em seguida faça o gráfico 

```python
platform_sales_sorted = platform_sales.sort_values('total_global_sales', ascending=False)
plt.figure(figsize=(12,5));
chart = platform_sales_sorted.total_global_sales.plot(kind='bar',facecolor='#AA0000', title='Global Sales by Platform');
chart.set_xlabel("Platform", fontsize=12);
chart.set_ylabel("Global Sales", fontsize=12);
```

## Gráfico de barras empilhado

### DataFrame - Vendas Regionais agrupadas por genêro

```python
genre = sales[[ "Genre", 'NA_Sales','EU_Sales','JP_Sales']].groupby(['Genre'], as_index=False).sum()
genre
```

### Drawing the stacked plot

```python
f, ax = plt.subplots(1, figsize=(12,5))

#salvar o nome das variaveis x
bar_l = genre['Genre']

ax.bar(bar_l, genre['NA_Sales'], label='NA', color='red')
ax.bar(bar_l, genre['EU_Sales'], bottom=genre['NA_Sales'], label='EU',color='blue')
#  o zip seria o mesmo que esta contido, que contem nas duas colunas Na_sales e EU_sales.
ax.bar(bar_l, genre['JP_Sales'], bottom=[i+j for i,j in zip(genre['NA_Sales'],genre['EU_Sales'])], \
       label='Japan',color='green')

ax.set_title("Vendas por Gênero e Região")
ax.set_xlabel("Gênero");
ax.set_ylabel("Vendas");

plt.xticks( rotation = 45, size = 14)

plt.legend(loc='upper right');
```

### Mais gráfico de barras 

```python
movie_scores = pd.read_csv('movie_scores.csv')
```

```python
# Create figure
plt.figure(figsize=(10, 5), dpi=300)

# Create bar plot
pos = np.arange(len(movie_scores['MovieTitle']))
width = 0.1
plt.bar(pos - width / 2, movie_scores['Tomatometer'], width, label='Tomatometer')
plt.bar(pos + width / 2, movie_scores['AudienceScore'], width, label='Audience Score')

# Specify ticks
plt.xticks(pos, rotation=30)
plt.yticks(np.arange(0, 101, 20))

# Get current Axes for setting tick labels and horizontal grid
ax = plt.gca()

# Set tick labels
ax.set_xticklabels(movie_scores['MovieTitle'])
ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])

# Add minor ticks for y-axis in the interval of 5
ax.set_yticks(np.arange(0, 100, 5), minor=True)

# Add major horizontal grid with solid lines
ax.yaxis.grid(which='major')

# Add minor horizontal grid with dashed lines
ax.yaxis.grid(which='minor', linestyle='--')

# Add title
plt.title('Movie comparison')

# Add legend
plt.legend()

# Show plot
plt.show()
```

### Vendas por região

```python
sales_by_region = [sales[:100]['NA_Sales'], 
                   sales[:100]['EU_Sales'], 
                   sales[:100]['JP_Sales'], 
                   sales[:100]['Global_Sales']]
```

```python
plt.figure(1, figsize=(12, 6))

regioes = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Global_Sales'] 

plt.boxplot(sales_by_region, labels = regioes, showmeans = True, whis = 1.5)

plt.xlabel("Região");
plt.ylabel("Vendas");

plt.xticks( rotation = 45, size = 20)
plt.yticks( size = 20)

plt.show;
```
